```python
df_m = df2.dropna(subset=["Municipality_kanji", "TradePrice", "Area"]).copy()
df_m["calc_unit_price"] = df_m["TradePrice"] / df_m["Area"]

muni_summary = df_m.groupby("Municipality_kanji").agg(
    count=("TradePrice", "count"),
    avg_trade_price=("TradePrice", "mean"),
    median_trade_price=("TradePrice", "median"),
    avg_unit_price=("calc_unit_price", "mean"),
    median_unit_price=("calc_unit_price", "median")
).reset_index()

muni_summary = muni_summary.sort_values("avg_unit_price", ascending=False)
muni_summary.head(20)
```

```python
# 구 / 시 / 마을 컬럼 초기화
df2["Ward"] = None
df2["City"] = None
df2["TownVillage"] = None

# 행정구분 컬럼
def area_type(name):
    if pd.isna(name):
        return None
    if name.endswith("区"):
        return "区"
    elif name.endswith("市"):
        return "市"
    elif name.endswith("町") or name.endswith("村"):
        return "町村"
    else:
        return "その他"

df2["AreaType"] = df2["Municipality_kanji"].apply(area_type)

# 값 채우기
df2.loc[df2["AreaType"] == "区", "Ward"] = df2["Municipality_kanji"]
df2.loc[df2["AreaType"] == "市", "City"] = df2["Municipality_kanji"]
df2.loc[df2["AreaType"] == "町村", "TownVillage"] = df2["Municipality_kanji"]

df2[["Municipality_kanji", "AreaType", "Ward", "City", "TownVillage"]].drop_duplicates().head()
```

```python
# ㎡ 단가
df2["unit_price"] = df2["TradePrice"] / df2["Area"]

# 건물 연식
df2["building_age"] = df2["Year"] - df2["BuildingYear"]
```

```python
master_cols = [
    # 지역 구조
    "AreaType", "Ward", "City", "TownVillage", "Municipality_kanji",

    # 가격
    "TradePrice", "Area", "unit_price",

    # 입지
    "TimeToNearestStation", "NearestStation",

    # 건물
    "BuildingYear", "building_age", "Structure", "FloorPlan", "Renovation",

    # 시계열
    "Year", "Quarter", "Period"
]

master_df = df2[master_cols].copy()

master_df.to_csv(
    "tokyo_master.csv",
    index=False,
    encoding="utf-8-sig"
)

print("저장완료")
```

```python
df_s = df2.dropna(subset=["TradePrice", "Area", "TimeToNearestStation"]).copy()
df_s["calc_unit_price"] = df_s["TradePrice"] / df_s["Area"]

bins = [-1, 5, 10, 15, 20, 999]
labels = ["0-5", "6-10", "11-15", "16-20", "21+"]

df_s["station_time_group"] = pd.cut(df_s["TimeToNearestStation"], bins=bins, labels=labels)

station_summary = (
    df_s.groupby("station_time_group", observed=False)
    .agg(
        count=("calc_unit_price", "count"),
        avg_unit_price=("calc_unit_price", "mean"),
        median_unit_price=("calc_unit_price", "median"),
        avg_trade_price=("TradePrice", "mean")
    )
    .reset_index()
)

station_summary
```

```python
station_summary.to_csv("tokyo_station_time.csv", index=False, encoding="utf-8-sig")
print("저장완료")
```

```python
# ㎡당 가격과 역거리/연식/면적 관계 보기

df_mid = df2.dropna(subset=["TradePrice", "Area", "TimeToNearestStation", "BuildingYear", "Year"]).copy()

df_mid["unit_price"] = df_mid["TradePrice"] / df_mid["Area"]
df_mid["building_age"] = df_mid["Year"] - df_mid["BuildingYear"]

df_mid[["unit_price", "Area", "TimeToNearestStation", "building_age"]].head()

df_mid[["unit_price", "Area", "TimeToNearestStation", "building_age"]].corr()
```

```python
mid_export = df_mid[[
    "Municipality_kanji", "DistrictName", "NearestStation", "Type",
    "TradePrice", "Area", "unit_price",
    "TimeToNearestStation", "BuildingYear", "building_age",
    "Structure", "FloorPlan", "Renovation",
    "Year", "Quarter"
]].copy()

mid_export.to_csv("tokyo_mid_clean.csv", index=False, encoding="utf-8-sig")
print("저장완료")
```

```python
# 연도별 가격추이

df_ts = df2.dropna(subset=["TradePrice", "Area", "Year", "Quarter"]).copy()
df_ts["unit_price"] = df_ts["TradePrice"] / df_ts["Area"]

ts_summary = df_ts.groupby(["Year", "Quarter"]).agg(
    count=("unit_price", "count"),
    avg_unit_price=("unit_price", "mean"),
    median_unit_price=("unit_price", "median"),
    avg_trade_price=("TradePrice", "mean")
).reset_index()

ts_summary = ts_summary.sort_values(["Year", "Quarter"])
ts_summary.head()
```

```python
ts_summary.to_csv("tokyo_time_series.csv", index=False)
print("저장완료")
```
